```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Task 1: Load and Explore the Dataset
#Choose a dataset in CSV format (for example, you can use datasets like the Iris dataset, a sales dataset, or any dataset of your choice).
#Load the dataset using pandas.
#Display the first few rows of the dataset using .head() to inspect the data.

try:
    df = pd.read_csv('farm_production_dataset.csv')
    print("✅ Dataset loaded successfully!")
    display(df.head())
except FileNotFoundError:
    print("❌ Error: File not found. Please check the file path.")
    
```

```python
#Explore the structure of the dataset by checking the data types and any missing values.
df.dtypes
```

```python
#Clean the dataset by either filling or dropping any missing values

print(df.dropna(inplace=True))
print(df.isnull().sum())
```

```python
#Task 2: Basic Data Analysis
#1. Compute the basic statistics of the numerical columns (e.g., mean, median, standard deviation) using .describe().

df.drop(columns=['REF_DATE']).describe()
```

```python
#2. Perform groupings on a categorical column (for example, species, region, or department) and compute the mean of a numerical column for each group.
df.groupby("Type of crop").agg({'Average farm price (dollars per tonne)': ['mean', 'std', 'median'],'Average yield (kilograms per hectare)': ['mean', 'std', 'median']})


```

```python
#3. Identify any patterns or interesting findings from your analysis.
"""--Lentils has the highest farm price per tonne
--Corn for silage has lowest farm price per tonne
--Tame hay has the highest yield
--Sunflower has the lowest yield"""
```

```python
#Task 3: Data Visualization
#1.Create at least four different types of visualizations:
#--Bar chart showing the comparison of a numerical value across categories (e.g., average petal length per species).
#2.Customize your plots with titles, labels for axes, and legends where necessary.
df.groupby('Type of crop')['Average farm price (dollars per tonne)'].mean().plot(kind='bar')
plt.xlabel('Type of crop')
plt.ylabel('Average farm price (dollars per tonne)')
plt.title('Average farm price of Type of crop')
plt.show()
```

```python
#--Line chart showing trends over time (for example, a time-series of sales data)
# Filter for crop
wheat_df = df[df['Type of crop'] == 'Wheat, all']

wheat_avg = wheat_df.groupby('REF_DATE')['Average farm price (dollars per tonne)'].mean()

plt.figure(figsize=(10, 6))
plt.plot(wheat_avg.index, wheat_avg.values, marker='o', linestyle='-', color='green')
plt.title('Average Farm Price of Wheat Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price (Dollars per Tonne)')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
#--Histogram of a numerical column to understand its distribution.
df_clean = df['Average farm price (dollars per tonne)'].dropna()
plt.figure(figsize=(8, 5))
plt.hist(df_clean, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Farm Price')
plt.xlabel('Price (Dollars per Tonne)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#--Scatter plot to visualize the relationship between two numerical columns (e.g., sepal length vs. petal length).
df.columns = df.columns.str.strip()
df_clean = df[['Average yield (kilograms per hectare)', 'Average farm price (dollars per tonne)']].dropna()
plt.figure(figsize=(8, 5))
plt.scatter(
    df_clean['Average yield (kilograms per hectare)'],
    df_clean['Average farm price (dollars per tonne)'],
    alpha=0.6, color='purple'
)
plt.title('Average Yield vs Average Farm Price')
plt.xlabel('Average Yield (kg/ha)')
plt.ylabel('Average Farm Price ($/tonne)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
